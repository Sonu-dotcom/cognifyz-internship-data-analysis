```python
price_range=df['Price range'].value_counts().sort_index()
```

```python
price_range
```

```python

plt.bar(price_range.index, price_range.values, color='skyblue', edgecolor='black', linewidth=1.2)
plt.title('Percentage of Restaurants in Each Price Range Category')
plt.xlabel('Price Range')
plt.ylabel('Percentage of Restaurants')
plt.xticks(rotation=0)
plt.show()
```

```python
percentage=(price_range/price_range.sum())*100
percentage
```
