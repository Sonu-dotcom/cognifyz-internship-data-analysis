```python
total_restaurants=len(df)
```

```python
total_restaurants
```

```python
online_delivery=df[df['Has Online delivery']=='Yes'].shape[0]
```

```python
online_delivery
```

```python
percentage=(online_delivery/total_restaurants)*100
percentage
```

```python
average_rating=df[df['Has Online delivery']=='Yes']['Aggregate rating'].mean()
average_rating
```

```python
average_rating_no_online=df[df['Has Online delivery']=='No']['Aggregate rating'].mean()
```

```python
average_rating_no_online
```

```python
import matplotlib.pyplot as plt
```

```python
labels=['average_rating','average_rating_no_online']
```

```python
plt.figure(figsize=(10, 6))
plt.bar(labels,average_rating, color=['skyblue', 'lightgreen'], edgecolor='black', linewidth=1.2)
plt.title('Average Ratings of Restaurants With and Without Online Delivery')
plt.xlabel('Online Delivery Option')
plt.ylabel('Average Rating')
plt.ylim(0,5)  
plt.show()
```
